<div class="card" dir="rtl">
  <h2 style="margin-bottom: 4px;">مثال</h2>
  <h4 style="margin-top: 0; color: #555; font-weight: normal;">یک ابزار ساده یادگیری ریاضی </h4>

  <p>
    این برنامه برای آزمون ریاضی یک دانش‌آموز کلاس اولی نوشته می‌شود. 
    دو عدد تک‌رقمی <code>number1</code> و <code>number2</code> (مثلاً ۷ و ۹) به‌طور تصادفی تولید شده و پرسش زیر را مطرح می‌کند:
  </p>

  <p style="direction:ltr; text-align:center; font-weight:bold; font-size:1.1em;">
    What is 7 + 9?
  </p>

  <p>
    پس از این‌که دانش‌آموز پاسخ خود را وارد کرد، پاسخ وارد شده با پاسخ واقعی مقایسه می‌شود و پیغامی مبنی بر 
    <span style="color:green; font-weight:bold;">true</span> یا 
    <span style="color:red; font-weight:bold;">false</span> بودن پاسخ چاپ می‌شود.
  </p>

  <div style="background-color:#f0f8ff; padding:10px; border-radius:8px; margin-bottom:15px;">
    <strong>برای تولید اعداد تصادفی صحیح در بازه 0 تا 9 از دستور زیر استقاده می شود.</strong>
  </div>

  <pre style="background-color:#272822; color:#f8f8f2; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
int num1 = (int)(Math.random() * 10); // returns a number from 0 to 9
  </pre>
</div>

```java
import java.util.Scanner;

public class AdditionQuiz {
    public static void main(String[] args) {
        int number1 = (int) (Math.random() * 10);
        int number2 = (int) (Math.random() * 10);
        
        // Creat a scanner
        Scanner input = new Scanner(System.in);

        System.out.print("What is " + number1 + " + " + number2 + "? ");
        int answer = input.nextInt();

        System.out.println(
            number1 + " + " + number2 + " = " + answer +" is " + (number1 + number2 == answer));
    }
}
```

  <h2 style="margin-bottom: 4px;">مثال</h2>
  <h4 style="margin-top: 0; color: #555; font-weight: normal;">محاسبه مساحت دایره با بررسی شعاع</h4>

  <p>
    برنامه‌ای بنویسید که مقدار شعاع دایره را از کاربر دریافت کند. اگر مقدار شعاع منفی نباشد، مساحت دایره محاسبه و چاپ شود.
  </p>

  <pre style="background-color:#fdf6e3; color:#333; padding:12px; border-radius:8px; direction:ltr; text-align:left;">

    area = radius * radius * PI;
    System.out.println("The area"
        + " for the circle of radius "
        + radius + " is " + area);
}
  </pre>

  <img src="img\page8f2.png" alt="example flowchart" style="width:100%; max-width:500px; display:block; margin:auto;">




</div>

```java
import java.util.Scanner;

public class CircleArea {
  public static void main(String[] args) {
    Scanner input = new Scanner(System.in);
    System.out.print("Enter radius: ");
    double radius = input.nextDouble();

    if (radius >= 0) {
      double area = radius * radius * Math.PI;
      System.out.println("The area for the circle of radius " + radius + " is " + area);
    }
  }
}
```

    برنامه‌ای بنویسید که یک عدد صحیح را از کاربر بگیرد. اگر عدد مضرب ۵ بود، پیغام <strong>HiFive</strong> و اگر مضرب ۲ بود، پیغام <strong>HiEven</strong> چاپ کند.

```java
import java.util.Scanner;

public class SimpleIfDemo {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);
        System.out.println("Enter an integer: ");
        int number = input.nextInt();

        if (number % 5 == 0)
            System.out.println("HiFive");

        if (number % 2 == 0)
            System.out.println("HiEven");
    }
}
```

  <h2 style="margin-bottom: 4px;">مثال </h2>

  <p>
    برنامه‌ای بنویسید که مقدار شعاع یک دایره را دریافت کند. اگر شعاع مقدار غیرمنفی داشته باشد، مساحت دایره محاسبه و چاپ شود. در غیر این‌صورت، پیام "Negative input" نمایش داده شود.
  </p>

```java
import java.util.Scanner;

public class CircleArea {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        System.out.print("Enter radius: ");
        double radius = input.nextDouble();

        if (radius >= 0) {
            double area = radius * radius * 3.14159;
            System.out.println("The area for the circle of radius " + radius + " is " + area);
        } else {
            System.out.println("Negative input");
        }
    }
}
```

<p style="background-color: #f9f9f9; padding: 10px; border-radius: 8px;">
  در روش اول از <strong>if‌های تودرتو (nested if)</strong> استفاده شده است؛ یعنی هر شرط درون بخش <code>else</code> شرط قبلی نوشته می‌شود و در نتیجه سطح تورفتگی (indent) افزایش می‌یابد.  
  از نظر عملکرد، نتیجه‌ی این روش با روش <strong>else if</strong> کاملاً <strong>یکسان</strong> است، فقط شکل نگارش متفاوت است و روش دوم خواناتر است.  
  همچنین چون در هر بخش تنها یک دستور وجود دارد، از آکولاد استفاده نشده است؛ اما اگر بیش از یک دستور در هر بخش داشتیم، باید آن‌ها را داخل آکولاد <code>{ }</code> قرار دهیم.

  <h2 style="margin-bottom: 4px;">خطای برنامه‌نویسی</h2>
  

  <p>
    قرار دادن نقطه ویرگول (;) بلافاصله بعد از شرط <code>if</code> باعث می‌شود بدنهٔ شرط عملاً «دستور خالی» باشد
    و بلوک آکولادیِ بعدی یک بلوک مستقل تلقی شود که همیشه اجرا می‌شود. این خطا نه خطای نحوی است و نه در زمان اجرا
    استثنا می‌دهد، به همین دلیل کشف آن دشوار است و صرفاً خروجی/منطق برنامه اشتباه می‌شود.
  </p>

  <p><strong>کد اشتباه:</strong></p>
  <pre style="background-color:#fdf6e3; color:#333; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
if (radius >= 0); // Empty statement ends the if condition
{
    area = radius * radius * PI; // This block runs regardless of the if condition
    System.out.println(
        "The area for the circle of radius " + radius + " is " + area);
}
  </pre>

  <p><strong>منطق خطا(آنچه واقعا اجرا میشود):</strong></p>
  <pre style="background-color:#fdf6e3; color:#333; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
// if statement has an empty body and does nothing
if (radius >= 0)
    ;  // Empty statement

// This block is independent of the if and always executes
{
    area = radius * radius * PI;
    System.out.println(
        "The area for the circle of radius " + radius + " is " + area);
}
  </pre>

  <p><strong>کد صحیح</strong></p>
  <pre style="background-color:#fdf6e3; color:#333; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
if (radius >= 0) { // Only run this block when the condition is true
    area = radius * radius * PI; // Calculate the area
    System.out.println(
        "The area for the circle of radius " + radius + " is " + area); // Print result
}
  </pre>


  <div class="slide-note">
    • <strong>دلیل بروز خطا:</strong> نقطه ویرگول بعد از if پایان دستور شرطی را اعلام می‌کند؛ بنابراین شرط فقط ارزیابی می‌شود و هیچ
    بدنه‌ای ندارد. آکولادهای بعدی یک بلوک جداگانه هستند که همیشه اجرا می‌شوند (چه شرط درست باشد چه غلط).<br>
    • <strong>راه‌حل:</strong> نقطه ویرگول را حذف کنید و همواره برای بدنهٔ if از آکولاد استفاده کنید — حتی اگر یک دستور باشد — تا از ابهام جلوگیری شود.<br>
  </div>

  <h2 style="margin-bottom: 4px;">مثال</h2>
  <h4 style="margin-top: 0; color: #555; font-weight: normal;">یک ابزار ساده یادگیری ریاضی </h4>
  <!-- توضیح عملکرد برنامه -->
  <p style="background-color: #f9f9f9; padding: 10px; border-radius: 8px;">
    این برنامه دو عدد تصادفی تولید می‌کند: <code>number1</code> و <code>number2</code>.  
    اگر <code>number1 > number2</code> باشد، یک سؤال تفریق نمایش داده می‌شود، مانند:
  </p>
  <p>نکته: تضمین می‌شود خروجی حتما مثبت است.</p>
  <!-- نمونه سؤال -->
  <pre style="background-color: #272822; color: #f8f8f2; padding: 10px; border-radius: 8px; direction: ltr; text-align: left;">
What is 9 - 2?
  </pre>

  <!-- توضیح پاسخ‌دهی -->
  <p style="background-color: #f9f9f9; padding: 10px; border-radius: 8px;">
    دانش‌آموز پاسخ را وارد می‌کند.  
    اگر پاسخ درست باشد، پیام <strong dir="ltr">"You are correct."</strong> نمایش داده می‌شود.  
    در صورت اشتباه،
     پیام <strong dir= "ltr">"Your answer is wrong."</strong> همراه با پاسخ صحیح نشان داده می‌شود.
  </p>
</div>

```java
import java.util.Scanner;

public class SubtractionQuiz {
    public static void main(String[] args) {
        // 1. Generate two random single-digit integers
        int number1 = (int)(Math.random() * 10);
        int number2 = (int)(Math.random() * 10);

        // 2. If number1 < number2, swap number1 with number2
        if (number1 < number2) {
            int temp = number1;
            number1 = number2;
            number2 = temp;
        }

        // 3. Prompt the student to answer "what is number1 - number2?"
        System.out.print("What is " + number1 + " - " + number2 + "? ");
        Scanner input = new Scanner(System.in);
        int answer = input.nextInt();

        // 4. Grade the answer and display the result
        if (number1 - number2 == answer)
            System.out.println("You are correct!");
        else
            System.out.println("Your answer is wrong.\n" + number1 + " - " + number2 + " should be " + (number1 - number2));
    }
}
```

  <h2 style="margin-bottom: 4px;">مثال</h2>
  <h4 style="margin-top: 0; color: #555; font-weight: normal;">
    جابجایی دو عدد با استفاده از عملگر XOR
  </h4>

  <p>
    برنامه‌ای بنویسید که دو عدد صحیح را از کاربر دریافت کند و بدون استفاده از متغیر کمکی، آن‌ها را با استفاده از عملگر XOR جابجا کند.
    <br><br>
    <br><br>
    به عنوان مثال، اگر ورودی‌ها به صورت زیر باشند:
    <br>عدد اول (a) = ۵
    <br>عدد دوم (b) = ۶
    <br>پس از اجرای برنامه، خروجی باید به صورت زیر باشد:
    <br>a = ۶
    <br>b = ۵
  </p>

  <p>پرسش: پس از خواندن کد بررسی کنید که عملیات انجام شده چگونه باعث جابه‌جایی دو عدد شده است؟</p>
</div>

```java
import java.util.Scanner;

public class SwapWithXOR {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        // Get two integers from the user
        System.out.print("Enter first number (a): ");
        int a = input.nextInt();

        System.out.print("Enter second number (b): ");
        int b = input.nextInt();

        // Display original values
        System.out.println("Before swap: a = " + a + ", b = " + b);

        // Swap using XOR
        a = a ^ b; // Step 1: a holds a ^ b
        b = a ^ b; // Step 2: b becomes original a
        a = a ^ b; // Step 3: a becomes original b

        // Display swapped values
        System.out.println("After swap: a = " + a + ", b = " + b);
    }
}
```

  <p>
    برنامه‌ای بنویسید که عددی را از کاربر دریافت کند و بررسی کند آیا:
    <br>۱. بر ۲ و ۳ بخش‌پذیر است.
    <br>۲. بر ۲ یا ۳ بخش‌پذیر است.

```java
import java.util.Scanner;

public class DivisibilityChecker {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);

        // Get input from user
        System.out.print("Enter an integer: ");
        int number = input.nextInt();

        // Check divisibility by both 2 and 3
        System.out.println("Is " + number + " divisible by 2 and 3? " +
            ((number % 2 == 0) && (number % 3 == 0)));

        // Check divisibility by either 2 or 3
        System.out.println("Is " + number + " divisible by 2 or 3? " +
            ((number % 2 == 0) || (number % 3 == 0)));

        // Check divisibility by 2 or 3, but not both
        System.out.println("Is " + number + " divisible by 2 or 3, but not both? " +
            ((number % 2 == 0) ^ (number % 3 == 0)));
    }
}
```

```java
import java.util.Scanner;

public class LeapYear {
    public static void main(String[] args) {
        // Create a Scanner to get user input
        Scanner input = new Scanner(System.in);
        System.out.print("Enter a year: ");
        int year = input.nextInt();

        // Check if the year is a leap year
        boolean isLeapYear =
            (year % 4 == 0 && year % 100 != 0) || (year % 400 == 0);

        // Display the result
        System.out.println(year + " is a leap year? " + isLeapYear);
    }
}
```

  <h2>عملگرهای بیتی (Bitwise Operators) در جاوا</h2>

  <p>
    در جاوا، <strong>عملگرهای بیتی</strong> مجموعه‌ای از عملگرها هستند که عملیات خود را بر روی داده‌های عددی در سطح <strong>بیت</strong> انجام می‌دهند.  
    این عملگرها تنها بر روی انواع داده صحیح مانند <code>int</code>، <code>long</code>، <code>short</code>، <code>char</code> و <code>byte</code> قابل استفاده‌اند.  
    تفاوت اصلی آن‌ها با عملگرهای منطقی (Logical) این است که نتیجه را بر اساس مقایسه و ترکیب بیت‌ها به دست می‌آورند، نه بر اساس درست و نادرست بودن یک شرط.
  </p>

  <p>
    هر عدد در حافظه به صورت یک رشته باینری (0 و 1) نمایش داده می‌شود.  
    عملگرهای بیتی این رشته باینری را بیت به بیت بررسی کرده و خروجی را بر اساس قوانین خود ایجاد می‌کنند.
  </p>

  <h3>مثال </h3>

  <h2 style="margin-bottom: 4px;">اولویت عملگرها</h2>

  <p>
    در این جدول، عملگرهای رایج در زبان‌های برنامه‌نویسی به ترتیب اولویت اجرا فهرست شده‌اند. این ترتیب به شما کمک می‌کند تا عبارات پیچیده را بهتر تحلیل و پیاده‌سازی کنید.
  </p>

  <pre style="background-color:#fdf6e3; color:#333; padding:12px; border-radius:8px; direction:ltr; text-align:left;">
1. var++, var-- : افزایش یا کاهش مقدار متغیر (پسوندی)
2. ++var, --var : افزایش یا کاهش مقدار متغیر (پیشوندی)
3. ! : عملگر منفی منطقی (NOT)
4. ~ : مکمل بیت‌ها (bitwise NOT)
5. (type) : تبدیل نوع داده (Casting)
6. *, /, % : ضرب، تقسیم و باقیمانده
7. +, - : جمع و تفریق
8. <<, >>, >>> : شیفت بیت‌ها به چپ یا راست
9. <, <=, >, >= : مقایسه کوچکتر، بزرگتر و مساوی
10. ==, != : بررسی برابری یا نابرابری
11. ^ : یا منطقی انحصاری (XOR)
12. & : و منطقی یا بیت‌به‌بیت (AND)
13. ^ : یا منطقی انحصاری بیت‌به‌بیت (XOR)
14. | : یا منطقی یا بیت‌به‌بیت (OR)
15. && : و شرطی (AND کوتاه‌شده)
16. || : یا شرطی (OR کوتاه‌شده)
17. ?: : عملگر شرطی (سه‌تایی)
18. =, +=, -=, *=, /=, %=, &=, ^=, |=, <<=, >>=, >>>= : عملگرهای انتساب
  </pre>

  <div class="slide-note">
    این لیست به شما کمک می‌کند تا بدانید کدام عملگرها ابتدا اجرا می‌شوند و چگونه می‌توان عبارات منطقی و ریاضی را به‌درستی ساختار داد.
    <br><br>
    <strong>تقدم عملگرها و شرکت‌پذیری:</strong>
    <ul style="margin-top: 8px;">
      <li>شرکت‌پذیری همه عملگرها دوعملوندی (به جز انتساب) از چپ به راست می‌باشد.</li>
      <li>شرکت‌پذیری عملگرها تک‌عملوندی افزایشی (پیشوندی و پسوندی) از راست به چپ می‌باشد.</li>
      <li>شرکت‌پذیری عملگرها شرطی <code>:?</code> از راست به چپ می‌باشد.</li>
    </ul>
  </div>

  <div class="slide-note" style="margin-top:16px; background:#eef9f2; padding:10px; border-radius:8px;">
    <strong>نکته درباره عملگر تقسیم در جاوا:</strong>

  <pre style="background-color:#fdf6e3; padding:10px; border-radius:8px; direction:ltr; text-align:left;"> 3 / 2 </pre>

  <p>هر دو عدد صحیح هستند، بنابراین تقسیم به‌صورت صحیح انجام شده و حاصل یک عدد صحیح برابر با یک است.</p>

  <pre style="background-color:#fdf6e3; padding:10px; border-radius:8px; direction:ltr; text-align:left;">3.0 /2 </pre>

  <pre style="background-color:#fdf6e3; padding:10px; border-radius:8px; direction:ltr; text-align:left;">int b = 3; int c = 2; float a = b / c;</pre>
  <p>ابتدا تقسیم صحیح انجام می‌شود و سپس نتیجه به نوع اعشاری تبدیل می‌گردد، بنابراین مقدار نهایی یک اعشاری خواهد بود.</p>

    <li>اگر کاربر پاسخ "Yes" بدهد، مقداری به متغیر <code>day</code> اضافه می‌شود (۱، ۲، ۴، ۸، ۱۶ بسته به شماره مجموعه).</li>
    <li>در پایان، مجموع این مقادیر روز واقعی تولد را مشخص می‌کند.</li>
    <li>این روش بر پایه نمایش دودویی (binary) روزها طراحی شده است.</li>
    <li>نتیجه با استفاده از <code>JOptionPane.showMessageDialog</code> به کاربر نمایش داده می‌شود.</li>

```java
import javax.swing.JOptionPane;
public class GuessBirthdayUsingConfirmationDialog {
    public static void main(String[] args) {
        String set1 =
            " 1  3  5  7\n" +
            " 9 11 13 15\n" +
            "17 19 21 23\n" +
            "25 27 29 31";

        String set2 =
            " 2  3  6  7\n" +
            "10 11 14 15\n" +
            "18 19 22 23\n" +
            "26 27 30 31";

        String set3 =
            " 4  5  6  7\n" +
            "12 13 14 15\n" +
            "20 21 22 23\n" +
            "28 29 30 31";

        String set4 =
            " 8  9 10 11\n" +
            "12 13 14 15\n" +
            "24 25 26 27\n" +
            "28 29 30 31";

        String set5 =
            "16 17 18 19\n" +
            "20 21 22 23\n" +
            "24 25 26 27\n" +
            "28 29 30 31";

        int day = 0;

        // Ask about each set and update 'day'
        int answer = JOptionPane.showConfirmDialog(null,"Is your birthday in these numbers?\n" + set1);
        if (answer == JOptionPane.YES_OPTION) day += 1;

        answer = JOptionPane.showConfirmDialog(null,"Is your birthday in these numbers?\n" + set2);
        if (answer == JOptionPane.YES_OPTION) day += 2;

        answer = JOptionPane.showConfirmDialog(null,"Is your birthday in these numbers?\n" + set3);
        if (answer == JOptionPane.YES_OPTION) day += 4;

        answer = JOptionPane.showConfirmDialog(null,"Is your birthday in these numbers?\n" + set4);
        if (answer == JOptionPane.YES_OPTION) day += 8;

        answer = JOptionPane.showConfirmDialog(null,"Is your birthday in these numbers?\n" + set5);
        if (answer == JOptionPane.YES_OPTION) day += 16;

        JOptionPane.showMessageDialog(null, "Your birthday is " + day + "!");
    }
}
```
