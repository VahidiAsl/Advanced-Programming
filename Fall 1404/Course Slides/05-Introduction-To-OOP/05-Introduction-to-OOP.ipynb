  <h3>یک شی دارای رفتار است</h3>
  
  <p>
      در شیوه‌های <strong>برنامه‌نویسی ساخت‌یافته (غیرشی‌گرا)</strong> داریم:
  </p>
     
  <ul style="margin-right: 2em;">
    <li>
      <strong>داده‌ها</strong> که به صورت <strong>انفعالی (غیر فعال)</strong> در برنامه استفاده می‌شوند.
    </li>  
    <li>
      <strong>توابع</strong>، که قادرند بر روی داده‌ها عملیاتی انجام دهند.
    </li>
    <li>
      با بزرگ‌تر شدن برنامه‌ها، <strong>مدیریت کد</strong> دشوارتر می‌شود.
    </li>
    <li>
      امکان بروز <strong>تداخل بین داده‌ها و توابع</strong> وجود دارد.
    </li>
    <li>
      افزودن قابلیت‌های جدید نیازمند <strong>تغییرات گسترده</strong> است.
    </li>
  </ul>

  <hr class="divider" />

  <p>
    در شیوه <strong>برنامه‌نویسی شی‌گر</strong> برنامه از <strong>اشیا</strong> ساخته می‌شود. 
    در درون هر شی <strong>داده‌ها (Object)</strong> و <strong>متدهای مربوطه</strong> قرار می‌گیرند. 
    این متدها بر روی <strong>داده‌های همان شی</strong> دستکاری انجام می‌دهند و داده‌های اشیای دیگر را تغییر نخواهند داد.
  </p>

  <p>
    در رابطه با <strong>برنامه‌نویسی شی‌گرا</strong> و <strong>اشیا</strong> یک سری نکات قابل ملاحضه می‌باشد:
  </p>

  <ul style="margin-right: 2em;">
    <li><strong>یک شیء فعال (active)</strong> است و قادر است کارهایی را انجام دهد.</li>
    <li><strong>یک شیء مسئول داده‌های مربوط به خودش</strong> است.</li>
    <li>می‌تواند <strong>داده‌های خود را برای دیگر اشیاء در معرض نمایش و استفاده قرار دهد</strong>.</li>
    <li>هر شیء <strong>هویت (Identity)</strong> مخصوص به خود را دارد و از دیگر اشیاء متمایز است.</li>
    <li>یک شیء می‌تواند دارای <strong>وضعیت (State)</strong> متفاوت در زمان‌های مختلف باشد.</li>
    <li><strong>رفتار (Behavior)</strong> هر شیء توسط <strong>متدهای آن</strong> مشخص می‌شود.</li>
    <li>اشیاء می‌توانند با یکدیگر <strong>تعامل (Interaction)</strong> داشته باشند.</li>
    <li>یک شیء می‌تواند از روی <strong>کلاس‌های از پیش تعریف‌شده</strong> ساخته شود.</li>
  </ul>

```java
class Circle {
    /** The radius of this circle */
    double radius; //data field

    /** Construct a circle object */
    Circle() {
    }

    /** Construct a circle object */
    Circle(double newRadius) {
        radius = newRadius;
    }

    /** Return the area of this circle */
    double getArea() { //method
        return radius * radius * 3.14159;
    }
}
```

  </p>

  <h3>ایجاد کلاس تست (TestDrive)</h3>
  <p>
    در دومین مرحله یک <strong>کلاس تست</strong> به نام <strong>DogTestDrive</strong> ساخته می‌شود.
  </p>
  <p>
    این کلاس شامل <strong>متد main</strong> است که نقطه شروع اجرای برنامه محسوب می‌شود.
    در این مرحله هنوز کدی داخل متد main قرار نمی‌دهیم و فقط ساختار کلی کلاس را آماده می‌کنیم.
  </p>
  <p>
    هدف از این کلاس تست، آزمایش کردن کلاس Dog و اطمینان از درستی عملکرد آن است.
  </p>

  <h3>ساخت شیء و استفاده از متغیرها و متدها</h3>
  <p>
    در این مرحله به سراغ کلاس <strong>DogTestDrive</strong> برمی‌گردیم
    و در داخل <strong>متد main</strong> یک <strong>شیء</strong> از کلاس Dog می‌سازیم.
  </p>
  <p>
    برای ساخت شیء از دستور <strong>new</strong> استفاده می‌کنیم
    و شیء ساخته‌شده را در یک متغیر به نام <strong>d</strong> ذخیره می‌کنیم.
  </p>
  <p>

    <bdi dir="ltr"><code>Book b = new Book();</code></bdi> 
    و 
    <bdi dir="ltr"><code>Book c = new Book();</code></bdi>،
    دو <strong>متغیر ارجاعی</strong> به نام‌های <strong>b</strong> و <strong>c</strong> ساخته می‌شوند.
  </p>

  <p dir="rtl">
    هر کدام از این متغیرها با استفاده از دستور

  </p>

  <p dir="rtl">

    <strong>تعداد اشیاء (Objects)</strong> نیز برابر با ۲ خواهد بود.
    نکته مهم این است که هر بار که از دستور 
    <bdi dir="ltr"><code>new</code></bdi> 
    استفاده می‌کنیم، یک شیء جدید در <strong>heap</strong> ساخته می‌شود، حتی اگر نوع آن یکسان باشد.

    <bdi dir="ltr"><code>Book d = c;</code></bdi>،

  </p>

  <p dir="rtl">

  <h3>دسترسی به اشیا</h3>

  <p dir="rtl" style="font-size:16px; line-height:1.8;">

  </p>

  <p dir="rtl" style="font-size:16px; line-height:1.8;">
    به همین ترتیب، در خط <strong>;()Circle yourCircle = new Circle</strong>
    یک <strong>متغیر ارجاعی</strong> دیگر به نام <strong>yourCircle</strong> ایجاد می‌شود
    که در ابتدا <span style="color:red;">no value</span> دارد،
    اما بخش راست یعنی <strong>()new Circle</strong>
    یک <strong>شیء جدید</strong> از کلاس <strong>Circle</strong> با مقدار پیش‌فرض
    (مثلاً <strong>۱٫۰</strong> یا <strong>۰٫۰</strong> بسته به سازنده) می‌سازد
    و مرجع آن به <strong>yourCircle</strong> داده می‌شود.
  </p>

    حالا هر <strong>متغیر ارجاعی</strong> به یک <strong>شیء جداگانه</strong> وصل است.

    در حالی که <strong>شیء</strong> مربوط به <strong>myCircle</strong> هیچ تغییری نمی‌کند
    و مقدار <strong>۵٫۰</strong> خودش را حفظ می‌کند.
  </p>

  <p dir="rtl" style="font-size:16px; line-height:1.8;">
    به طور خلاصه:
    <strong>اعلان متغیر</strong> فقط یک <strong>ریموت خالی</strong> می‌سازد،
    <strong>new</strong> یک <strong>شیء جدید</strong> ایجاد می‌کند
    و عملگر <strong>=</strong> آن‌ها را به هم متصل می‌کند؛

    که <strong>متغیر ارجاعی</strong> به آن اشاره می‌کند.
  </p>

  <h3>مقادیر پیش‌فرض برای فیلدهای داده‌ای</h3>

<p dir="rtl">
اگر یک متغیر از نوع <strong>ارجاعی (reference type)</strong> تعریف شود اما به هیچ شیئی <strong>ارجاع داده نشود</strong>، مقدار آن به طور پیش‌فرض <strong>null</strong> خواهد بود. برای مثال در قطعه کد زیر:<br><br>
<bdi dir="ltr"><code>Circle myCircle;&lrm;</code></bdi><br><br>
متغیر <strong>myCircle</strong> فقط اعلان شده است و هنوز به هیچ شیئی متصل نیست؛ بنابراین مقدار آن <strong>null</strong> است، یعنی هیچ آدرس/مرجعی به یک شیء واقعی در حافظه ندارد.
</p>

<p dir="rtl">
<strong>مقادیر پیش‌فرض در جاوا:</strong><br>
- برای فیلد داده‌ای از نوع <strong>ارجاعی</strong>: مقدار <strong>null</strong><br>
- برای نوع <strong>عددی</strong>: مقدار <strong>0</strong><br>
- برای نوع <strong>بولین</strong>: مقدار <strong>false</strong><br>
- برای نوع <strong>کاراکتری</strong>: مقدار <strong>'\u0000'</strong><br><br>
با این حال، <strong>جاوا برای یک متغیر محلی درون یک متد مقدار پیش‌فرض در نظر نمی‌گیرد</strong> و اگر قبل از استفاده مقداردهی نشود، باعث خطا خواهد شد.
</p>


<hr class="divider"  />


</div>




```java
class Student {
    Char name;
    int age;
    boolean isScienceMajor;
    String gender;
}
```

```java
public class Test {
    public static void main(String[] args) {
        Student student = new Student();
        System.out.println("name? " + student.name);
        System.out.println("age? " + student.age);
        System.out.println("isScienceMajor? " + student.isScienceMajor);
        System.out.println("gender? " + student.gender);
    }
}
Test.main(new String[5])
```

  <p dir="rtl">
    این عکس‌ها تفاوت رفتار <strong>انتساب در انواع اصلی (Primitive)</strong> و
    <strong>انواع ارجاعی (Object)</strong> را نشان می‌دهند.
  </p>

  <p dir="rtl">
    در بخش چپ، با انتساب
    <bdi><code>i = j</code></bdi>
    مقدار عددی متغیر <strong>j</strong> مستقیماً در <strong>i</strong> کپی می‌شود؛
    بنابراین بعد از عمل، <strong>i</strong> مقدار جدید (مثلاً <strong>2</strong>) را دارد

  <p dir="rtl">
    اما در بخش راست، با انتساب
    <bdi><code>c1 = c2</code></bdi>
    تنها <strong>مرجع</strong> <strong>شیء</strong> کپی می‌شود، نه خود <strong>شیء</strong>;
    در نتیجه هر دو متغیر <strong>c1</strong> و <strong>c2</strong>
    به یک <strong>شیء مشترک</strong> اشاره می‌کنند
    (در تصویر شیئی با <strong>radius = 9</strong>).
  </p>

  <p dir="rtl">
    <strong>شیء</strong> قبلیِ <strong>c1</strong>
    (مثلاً با <strong>radius = 5</strong>)
    اگر مرجعی به آن وجود نداشته باشد،
    توسط <strong>Garbage Collector</strong> حذف خواهد شد.
  </p>

  <p dir="rtl">
    به این ترتیب، در <strong>انواع اصلی</strong> «<strong>کپی مقدار</strong>» انجام می‌شود،
    اما در <strong>انواع ارجاعی</strong> «<strong>کپی مرجع</strong>» و اشتراک در یک <strong>شیء</strong> اتفاق می‌افتد.
  </p>

  <h3>جمع‌آوری زباله (Garbage Collection)</h3>

  <p dir="rtl">
    همان‌طور که در <strong>توضیحات قبلی</strong> و <strong>عکس‌های قبلی</strong> دیدیم،
    پس از دستور انتساب <bdi><code>c1 = c2</code></bdi>،
    متغیر <strong>c1</strong> به همان <strong>شیء</strong>ی اشاره می‌کند که توسط <strong>c2</strong> مورد ارجاع قرار گرفته است.
  </p>

  <p dir="rtl">
    در این حالت، <strong>شیء</strong>ی که قبلاً توسط <strong>c1</strong> مورد ارجاع قرار می‌گرفت
    و دیگر هیچ <strong>متغیر ارجاعی</strong> به آن اشاره ندارد، بلااستفاده باقی می‌ماند.
    به چنین شیئی اصطلاحاً <strong>زباله</strong> گفته می‌شود.
    این زباله‌ها در زبان <strong>جاوا</strong> به صورت خودکار توسط <strong>JVM</strong> مدیریت و جمع‌آوری
    (<strong>Garbage Collection</strong>) می‌شوند.
  </p>

  <p dir="rtl">
    <strong>Garbage Collection</strong> در <strong>جاوا</strong> یکی از مهم‌ترین ویژگی‌ها است
    که باعث می‌شود مدیریت حافظه ساده‌تر شود.
    برخلاف زبان‌هایی مثل <bdi>C++</bdi> یا <bdi>C</bdi> که برنامه‌نویس باید حافظه را به صورت دستی آزاد کند،
    در <strong>جاوا</strong> این کار به طور خودکار توسط <strong>JVM</strong> انجام می‌شود.
  </p>

  <p dir="rtl">
    زمانی که هیچ <strong>متغیر ارجاعی</strong> به یک <strong>شیء</strong> اشاره نکند،
    آن شیء <strong>غیرقابل‌دسترسی</strong> محسوب می‌شود
    و <strong>Garbage Collector</strong> آن را آزاد می‌کند.
    این فرآیند باعث جلوگیری از <strong>Memory Leak</strong> (نشت حافظه)
    شده و استفاده بهینه‌تری از منابع سیستم به همراه دارد.
  </p>

  <p dir="rtl">
    البته باید توجه داشت که اجرای <strong>Garbage Collector</strong>
    بر اساس زمان‌بندی خاص <strong>JVM</strong> است
    و الزاماً در لحظه‌ای که ما انتظار داریم اجرا نمی‌شود.
    این موضوع می‌تواند در برخی موارد باعث کاهش سرعت برنامه برای مدت کوتاهی شود.
  </p>

  <p dir="rtl">
    به همین دلیل، در طراحی برنامه‌های بزرگ باید دقت داشت که
    تعداد زیادی <strong>شیء</strong> غیرضروری ایجاد نشود
    تا بار اضافی بر روی <strong>Garbage Collector</strong> تحمیل نگردد.
  </p>

  <p dir="rtl">
    <strong>نکته:</strong> اگر مطمئن هستید یک <strong>شیء</strong> دیگر مورد نیاز نیست،
    می‌توانید <strong>متغیر ارجاعی</strong> کنترل‌کنندهٔ آن را به مقدار <strong>null</strong> تنظیم کنید؛ مانند:
  </p>

  <p dir="rtl" style="text-align:center;">
    <bdi dir="ltr"><code>c1 = null;</code></bdi>
  </p>

  <p dir="rtl">
    با این کار، وقتی هیچ مرجع دیگری به آن <strong>شیء</strong> وجود نداشته باشد،
    <strong>JVM</strong> در فرآیند <strong>Garbage Collection</strong>
    آن را به‌طور خودکار آزاد می‌کند.
  </p>

<h3>متغیرهای نمونه (Instance variables)</h3>
  <p>
    متغیرهای نمونه داخل یک <strong>class</strong> تعریف می‌شوند، اما داخل متدها قرار ندارند.  
    هر وقت از آن کلاس یک <strong>شیء (object)</strong> بسازیم، این متغیرها برای آن شیء ایجاد می‌شوند و مقدار مخصوص به خودشان را دارند.  
    به همین دلیل به آن‌ها می‌گوییم «حالت یا state شیء».
  </p>
  <p>
    این متغیرها روی حافظه‌ی <strong>heap</strong> ساخته می‌شوند و تا وقتی شیء وجود دارد (و توسط <strong>Garbage Collector</strong> پاک نشده) باقی می‌مانند.  
    دسترسی به آن‌ها از طریق شیء انجام می‌شود، مثلاً:  
    <code dir="ltr">obj.field</code>
  </p>
  <p>
    این متغیرها می‌توانند سطح دسترسی‌های مختلف داشته باشند (مثل <strong>public / private</strong>) و حتی مقدار اولیه هم داشته باشند.  
    تفاوت اصلی آن‌ها با متغیرهای <strong>static</strong> این است که متغیرهای نمونه برای هر شیء یک نسخه جداگانه می‌سازند، در حالی که متغیر <strong>static</strong> فقط یک نسخه مشترک برای کل کلاس دارد.
  </p>
  <hr class="divider"  />
  <h3>متدهای نمونه (Instance methods)</h3>
  <p>
    متدهای نمونه عملیاتی هستند که روی یک شیء مشخص اجرا می‌شوند.  
    این متدها به متغیرهای نمونه همان شیء دسترسی مستقیم دارند و می‌توانند با کلیدواژه <strong>this</strong> به اعضای همان شیء اشاره کنند.
  </p>
  <p>
    برای استفاده از آن‌ها باید متد را روی یک شیء صدا بزنیم، مثلاً:  
    <code dir="ltr">obj.doSomething()</code>  
    اگر شیئی وجود نداشته باشد، این متدها را نمی‌توان از زمینه <strong>static</strong> فراخوانی کرد.
  </p>
  <p>
    متدهای نمونه کاربردهای زیادی دارند، مثلاً:  
    <strong>encapsulation</strong> (مثل <strong>getter/setter</strong>)،  
    بازنویسی متدها (<strong>override</strong>) در کلاس‌های فرزند،  
    و پیاده‌سازی <strong>polymorphism</strong> در برنامه‌نویسی شیءگرا.
  </p>

  <li>
    این مثال، به یک متغیر کلاسی <strong>numberOfObjects</strong> یک واحد اضافه می‌کند تا تعداد اشیای ساخته‌شده از کلاس <strong>Circle</strong> را حساب کند.

```java
public class Circle2 {
  /** The radius of the circle */
  double radius;

  /** The number of the objects created */
  static int numberOfObjects = 0;

  /** Construct a circle with radius 1 */
  Circle2() {
    radius = 1.0;
    numberOfObjects++;
  }

  /** Construct a circle with a specified radius */
  Circle2(double newRadius) {
    radius = newRadius;
    numberOfObjects++;
  }

  /** Return numberOfObjects */
  static int getNumberOfObjects() {
    return numberOfObjects;
  }

  /** Return the area of this circle */
  double getArea() {
    return radius * radius * Math.PI;
  }
}
```

```java
public class TestCircle2 {
  /** Main method */
  public static void main(String[] args) {
    System.out.println("Before creating objects");
    System.out.println("The number of Circle objects is " + Circle2.numberOfObjects);

    // Create c1
    Circle2 c1 = new Circle2();

    // Display c1 BEFORE c2 is created
    System.out.println("\nAfter creating c1");
    System.out.println("c1: radius (" + c1.radius +  ") and number of Circle objects (" + c1.numberOfObjects + ")");

    // Create c2
    Circle2 c2 = new Circle2(5);

    // Modify c1
    c1.radius = 9;

    // Display c1 and c2 AFTER c2 was created
    System.out.println("\nAfter creating c2 and modifying c1");
    System.out.println("c1: radius (" + c1.radius + ") and number of Circle objects (" + c1.numberOfObjects + ")");
    System.out.println("c2: radius (" + c2.radius + ") and number of Circle objects (" + c2.numberOfObjects + ")");
  }
}

TestCircle2.main(new String[5])
```

  <p>
    در <strong>برنامه‌نویسی شیءگرا</strong>، هر <strong>متغیر</strong> یا <strong>متد</strong> می‌تواند

  <h3>مثالی از کپسوله‌بندی فیلد داده‌ای</h3>
  
  <div class="images" style="text-align:center;">
    <img src="img/Picture61.png" alt="Code Explanation">
  </div>

  <hr class="divider" />

  <p>
    در زبان <strong>Java</strong> یکی از اصول مهم <strong>برنامه‌نویسی شیءگرا</strong>، 
    <strong>کپسول‌بندی (Encapsulation)</strong> است.  
    کپسول‌بندی یعنی این‌که <strong>فیلدهای داده‌ای (Data Fields)</strong> یا همان <strong>متغیرهای کلاس</strong>، 
    به صورت <strong>private</strong> تعریف شوند تا مستقیماً از بیرون کلاس قابل دسترسی نباشند.
  </p>

  <p>
    در عوض، برای <strong>دسترسی</strong> یا <strong>تغییر دادن مقدار</strong> آن‌ها از متدهای 
    <strong>getter</strong> و <strong>setter</strong> استفاده می‌کنیم.  
    به این ترتیب، داده‌ها درون کلاس <strong>ایمن‌تر</strong> هستند و برنامه‌نویس می‌تواند کنترل کند 
    که چگونه مقادیر <strong>خوانده</strong> یا <strong>تغییر داده</strong> شوند.
  </p>

  <p>
    مثلاً در مثال <strong>Circle</strong>، فیلد <strong>radius</strong> به صورت <strong>private</strong> تعریف شده است.  
    برای خواندن مقدار آن از متد <bdi>getRadius()</bdi> و برای تغییر مقدار آن از متد 
    <bdi>setRadius(double radius)</bdi> استفاده می‌شود.  
    این روش هم باعث <strong>افزایش امنیت داده</strong> می‌شود و هم کد را <strong>منظم‌تر</strong> و <strong>قابل‌نگهداری‌تر</strong> می‌کند.
  </p>

  <hr class="divider" />
</div>


```java
public class Circle3 {
    /** The radius of the circle */
    private double radius = 1;

    /** The number of the objects created */
    private static int numberOfObjects = 0;

    /** Construct a circle with radius 1 */
    public Circle3() {
        numberOfObjects++;
    }

    /** Construct a circle with a specified radius */
    public Circle3(double newRadius) {
        radius = newRadius;
        numberOfObjects++;
    }

    /** Return radius */
    public double getRadius() {
        return radius;
    }

    /** Set a new radius */
    public void setRadius(double newRadius) {
        radius = (newRadius >= 0) ? newRadius : 0;
    }

    /** Return numberOfObjects */
    public static int getNumberOfObjects() {
        return numberOfObjects;
    }

    /** Return the area of this circle */
    public double getArea() {
        return radius * radius * Math.PI;
    }
}
```

```java
public class TestCircle3 {
    /** Main method */
    public static void main(String[] args) {
        // Create a Circle with radius 5.0
        Circle3 myCircle = new Circle3(5.0);

        System.out.println("The area of the circle of radius "
                + myCircle.getRadius() + " is " + myCircle.getArea());

        // Increase myCircle's radius by 10%
        myCircle.setRadius(myCircle.getRadius() * 1.1);
        System.out.println("The area of the circle of radius "
                + myCircle.getRadius() + " is " + myCircle.getArea());
    }
}

TestCircle3.main(new String[5])
```

  <h3>ارسال اشیا به متدها</h3>

  <p>
    در <strong>Java</strong> دو شیوهٔ اصلی برای <strong>ارسال پارامتر</strong> وجود دارد.
  </p>

  <p>
    در حالت <strong>ارسال با مقدار (Pass by Value)</strong> برای 
    <strong>انواع اولیه (primitive)</strong> مانند <strong>int</strong> یا <strong>double</strong>، 
    یک <strong>کپی از مقدار</strong> به متد داده می‌شود.  
    هر تغییری که داخل متد روی پارامتر انجام شود، روی متغیر اصلی اثر نمی‌گذارد.
  </p>

  <p>
    در حالت <strong>ارسال با مقدارِ ارجاع (Reference Value)</strong> برای 
    <strong>اشیا و آرایه‌ها</strong>، در واقع <strong>کپیِ ارجاع</strong> (آدرس شیء) به متد داده می‌شود.  
    به همین دلیل اگر <strong>ویژگی‌های شیء</strong> داخل متد تغییر کنند، 
    این تغییرات روی <strong>شیء اصلی</strong> نیز اعمال خواهد شد، 
    چون هر دو به یک محل در حافظه اشاره می‌کنند.
  </p>

  <p>
    با این حال، اگر داخل متد پارامتر ارجاعی دوباره به یک <strong>شیء جدید</strong> انتساب داده شود، 
    این تغییر فقط در همان متد باقی می‌ماند و <strong>مرجع بیرونی</strong> همچنان به شیء قبلی اشاره می‌کند.
  </p>

  <p>
    این رفتار باعث می‌شود که تغییر مستقیم روی <strong>حالت شیء</strong> از متد قابل مشاهده باشد، 
    اما تغییر <strong>مرجع</strong> آن تنها در محدودهٔ متد اثر کند.
  </p>

  <hr class="divider" />
</div>


```java
public class TestPassObject {
    /** Main method */
    public static void main(String[] args) {
        // Create a Circle object with radius 1
        Circle3 myCircle = new Circle3(1);

        // Print areas for radius 1, 2, 3, 4, and 5.
        int n = 5;
        printAreas(myCircle, n);

        // See myCircle.radius and times
        System.out.println("\n" + "Radius is " + myCircle.getRadius());
        System.out.println("n is " + n);
    }

    /** Print a table of areas for radius */
    public static void printAreas(Circle3 c, int times) {
        System.out.println("Radius \t\t" + "Area");
        while (times >= 1) {
            System.out.println(c.getRadius() + "\t\t" + c.getArea());
            c.setRadius(c.getRadius() + 1);
            times--;
        }
    }
}
TestPassObject.main(new String[5])
```

<p> در زبان <strong>Java</strong> می‌توان <strong>آرایه‌ای از اشیا</strong> ایجاد کرد. دستور زیر یک آرایه از نوع <strong>Circle</strong> تعریف می‌کند که ظرفیت ۱۰ خانه دارد: </p> 

<pre dir="ltr" style="text-align:left; border:1px solid #ccc; padding:10px; background:#f9f9f9; border-radius:5px; color:black; font-family:Consolas, 'Courier New', monospace;">

<p> این دستور یک <strong>آرایه</strong> ایجاد می‌کند، اما توجه داشته باشید که در این لحظه <strong>۱۰ شی از Circle ساخته نمی‌شود</strong، بلکه تنها ۱۰ خانه برای <strong>ارجاع‌ها (references)</strong> به اشیای Circle ایجاد می‌گردد. به بیان دیگر، هر خانه از این آرایه در ابتدا مقدار <strong>null</strong> دارد تا زمانی که یک شی Circle واقعی به آن اختصاص داده شود. </p> 

<p>نکات مهم:</p> 
<ul style="margin-right: 2em;"> 
<li> <strong>آرایه‌ای از اشیا</strong> در واقع <strong>آرایه‌ای از متغیرهای ارجاعی</strong> است. یعنی هر عنصر آرایه می‌تواند به یک شی خاص در حافظه <strong>Heap</strong> اشاره کند. </li>

<li> خود <strong>circleArray</strong> یک ارجاع به کل آرایه است، یعنی متغیری که در <strong>Stack</strong> نگهداری شده و به فضایی در <strong>Heap</strong> اشاره می‌کند که شامل ۱۰ خانه (برای ۱۰ ارجاع) است. </li> 
<li> هر عنصر مانند <strong>circleArray[1]</strong> یک ارجاع به یک شی از نوع <strong>Circle</strong> است. تا زمانی که شی جدیدی به آن تخصیص داده نشود، مقدار آن <strong>null</strong> باقی می‌ماند. </li>

<ul style="margin-right: 2em;">

<pre dir="ltr" style="text-align:left; border:1px solid #ccc; padding:10px; background:#f9f9f9; border-radius:5px; color:black; font-family:Consolas, 'Courier New', monospace;">
for (int i = 0; i < circleArray.length; i++) {
    circleArray[i] = new Circle();
}
</pre>

<li> در سطح مفهومی، می‌توان گفت که در اینجا دو سطح ارجاع داریم:

<li>یک ارجاع برای کل آرایه (<strong>circleArray</strong>).</li> 
<li>چند ارجاع داخل آرایه که هر کدام می‌توانند به یک شی Circle در حافظه Heap اشاره کنند.</li> 
</ul> 
</li> 
</ul>

```java
public class TotalArea {
    /** Main method */
    public static void main(String[] args) {
        // Declare circleArray
        Circle3[] circleArray;

        // Create circleArray
        circleArray = createCircleArray();

        // Print circleArray and total areas of the circles
        printCircleArray(circleArray);
    }

    /** Create an array of Circle objects */
    public static Circle3[] createCircleArray() {
        Circle3[] circleArray = new Circle3[5];

        for (int i = 0; i < circleArray.length; i++) {
            circleArray[i] = new Circle3(Math.random() * 100);
        }
        // Return Circle array
        return circleArray;
    }

    /** Print an array of circles and their total area */
    public static void printCircleArray(Circle3[] circleArray) {
        System.out.printf("%-30s%-15s\n", "Radius", "Area");
        for (int i = 0; i < circleArray.length; i++) {
            System.out.printf("%-30f%-15f\n",
                circleArray[i].getRadius(),
                circleArray[i].getArea());
        }

        System.out.println("--------------------------------------------");

        // Compute and display the result
        System.out.printf("%-30s%-15f\n", "The total areas of circles is",
            sum(circleArray));
    }

    /** Add circle areas */
    public static double sum(Circle3[] circleArray) {
        // Initialize sum
        double sum = 0;

        // Add areas to sum
        for (int i = 0; i < circleArray.length; i++)
            sum += circleArray[i].getArea();

        return sum;
    }
}
```
