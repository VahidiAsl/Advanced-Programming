<div class="card" dir="rtl">
  <h2>نوشتن یک برنامه ساده در جاوا</h2>
  <p>
    برای شروع کار با زبان <strong>جاوا</strong>، یک برنامه ساده می‌نویسیم که تنها وظیفه‌اش چاپ یک رشته در خروجی باشد. 
    این برنامه شامل یک کلاس و متد <code>main</code> است که نقطه آغاز اجرای هر برنامه جاوا محسوب می‌شود.
  </p>
  <ul>
    <li>برای تعریف کلاس از کلیدواژه <code>public class</code> استفاده کرده و سپس نام کلاس را مشخص می‌کنیم.</li>
    <li>تمامی دستورات کلاس باید در میان آکولاد <code>{ }</code> نوشته شوند.</li>
    <li>کلاس باید شامل متد <code>main</code> باشد؛ این متد نقطه شروع اجرای برنامه است.</li>
    <li>ساختار متد اصلی به صورت <code>public static void main(String[] args)</code> تعریف می‌شود.</li>
    <li>درون متد <code>main</code> می‌توانیم با دستور <code dir="ltr">System.out.println()</code> متن دلخواه را چاپ کنیم.</li>
    <li>پس از هر دستور باید از <code>;</code> برای پایان دستور استفاده شود.</li>
  </ul>
  <div class="slide-note">
    <strong>نکته:</strong> هر فایل جاوا باید تنها شامل یک کلاس عمومی (public) باشد و نام فایل دقیقاً با نام همان کلاس یکسان ذخیره گردد.
  </div>
  <hr class="divider" />
</div>


```java
public class First {
    public static void main(String[] args) {
        System.out.println("Salam!!!");
    }
}
```

    <li>هیچ محدودیتی روی طول شناسه وجود ندارد.</li>
    <li>انتخاب نام‌های خوانا و معنادار برای شناسه‌ها باعث درک بهتر کد می‌شود.</li>

  <h2>برنامه‌نویسی با ورودی کاربر</h2>
  <p>
    در این بخش، قصد داریم با استفاده از کلاس <strong>Scanner</strong> از کاربر ورودی بگیریم و آن را پردازش کنیم.
    برای این کار، دو برنامه ساده اما مهم می‌نویسیم:
  </p>
  <ol>
    <li>
      <strong>محاسبه مساحت دایره:</strong> شعاع دایره توسط کاربر وارد می‌شود و برنامه مساحت دایره را محاسبه کرده و نمایش می‌دهد.
    </li>
    <li>
      <strong>تبدیل ثانیه به دقیقه و ثانیه:</strong> تعداد ثانیه‌ای که کاربر وارد می‌کند به دقیقه و ثانیه باقی‌مانده تبدیل و نمایش داده می‌شود.
    </li>
  </ol>
  <p>
  این تمرین‌ها کمک می‌کنند تا با نحوهٔ دریافت داده از کاربر و انجام محاسبات پایه در جاوا آشنا شوید.
  </p>
  <hr class="divider"/>
</div>


```java
import java.util.Scanner; // Scanner is in the java.util package

public class ComputeAreaWithConsoleInput {
  public static void main(String[] args) {
    // Create a Scanner object
    Scanner input = new Scanner(System.in);

    // Prompt the user to enter a radius
    System.out.print("Enter a number for radius: ");
    double radius = input.nextDouble();

    // Compute area
    double area = radius * radius * 3.14159;

    // Display result
    System.out.println("The area for the circle of radius " + radius + " is " + area);
  }
}
```

```java
import java.util.Scanner;

public class DisplayTime {
  public static void main(String[] args) {
    Scanner input = new Scanner(System.in);
    // Prompt the user for input
    System.out.print("Enter an integer for seconds: ");
    int seconds = input.nextInt();

    int minutes = seconds / 60; // Find minutes in seconds
    int remainingSeconds = seconds % 60; // Seconds remaining
    System.out.println(seconds + " seconds is " + minutes + " minutes and " + remainingSeconds + " seconds");
  }
}
```

  <pre dir="ltr" align="left"><code dir="ltr">double d2 = 100.2d;
</code></pre>

  <h2>تبدیل نوع داده‌ای (Type Conversion)</h2>
  <p>اگر دو عملوند از انواع مختلف باشند، جاوا به طور خودکار نوع مناسب را انتخاب می‌کند:</p>
  <ol>
    <li>اگر یکی <code>double</code> باشد ← دیگری هم به <code>double</code> تبدیل می‌شود.</li>
    <li>اگر یکی <code>float</code> باشد ← دیگری هم به <code>float</code> تبدیل می‌شود.</li>
    <li>اگر یکی <code>long</code> باشد ← دیگری هم به <code>long</code> تبدیل می‌شود.</li>
    <li>در غیر این صورت هر دو به <code>int</code> تبدیل می‌شوند.</li>
  </ol>

  <h3>مثال:</h3>
  <pre dir="ltr" align="left"><code>byte i = 100;
long k = i * 3 + 4;          // long ارتقا به
double d = i * 3.1 + k / 2;  // double ارتقا به 
</code></pre>
