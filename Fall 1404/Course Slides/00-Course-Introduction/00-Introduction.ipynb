  <h2>چرا جاوا</h2>


  <p>
    علاوه بر یادگیری ساختار و ویژگی‌های جاوا، با مفاهیم <strong>برنامه‌نویسی شیء‌گرا</strong> (OOP) که محبوب‌ترین متدولوژی توسعه نرم‌افزار است نیز آشنا خواهید شد.
  </p>
  <p>
    در ادامه، نگاهی بر برنامه‌نویسی در <strong>فریم‌وُرک فلاتر</strong> نیز خواهیم داشت تا بتوانید از دانش خود در توسعه اپلیکیشن‌ها استفاده کنید.
  </p>

  <div class="images-big">
    <img src="img/java.svg" alt="Java Logo" />
    <img src="img/flutter.jpg" alt="Flutter Logo" />
  </div>

  <p>
    در اکوسیستم نرم‌افزار امروز، <strong>جاوا</strong> همچنان بین زبان‌های پراستفاده قرار دارد: بر اساس نظرسنجی سال ۲۰۲۵ استک‌اُورفلو، حدود <strong>۲۹–۳۰٪</strong> از توسعه‌دهندگان حرفه‌ای در یک سال گذشته با جاوا کار کرده‌اند، و جاوا در شاخص TIOBE نیز در جمع ۴ زبان برتر است.
  </p>

  <p>
    جاوا زبان رسمی توسعه اندروید می‌باشد. بسیاری از نرم‌افزارهای تحت وب client-server به جاوا نوشته می‌شوند و میلیون‌ها برنامه‌نویس جاوا در این حوزه فعال هستند.
  </p>

  <p>
    از منظر بازار، هزینه مصرف‌کننده در اپ‌های موبایل در سال ۲۰۲۴ حدود <strong>۱۵۰ میلیارد دلار</strong> بوده (۱۳٪ رشد نسبت به ۲۰۲۳) که نشان می‌دهد فرصت اقتصادی اکوسیستم اپ بسیار بزرگ است؛ بخش قابل‌توجهی از این اپ‌ها به‌صورت بومی یا هیبرید برای اندروید توسعه می‌یابند.
  </p>

  <p>
    در سمت سرور، جاوا همچنان ستون فقرات بسیاری از سامانه‌های وب enterprise و کلاینت–سرور است (به‌ویژه با اکوسیستم Spring)، و به‌دلیل پایداری ABI/JVM و ابزار بالغ، برای سرویس‌های در مقیاس سازمانی انتخاب رایجی محسوب می‌شود.
  </p>

  <h3>Java Development Kit (JDK)</h3>
  <p>
    برای نوشتن، <strong>کامپایل</strong> و <strong>اجرای</strong> برنامه‌های جاوا به <strong>JDK</strong> نیاز داریم؛
    داشتن <strong>JRE</strong> به‌تنهایی فقط اجرای برنامه را ممکن می‌کند.
  </p>
  <ul>
   <li><strong>Development Tools</strong> — ابزارهای توسعه مانند <code>javac</code> (کامپایل)، <code>java</code> (اجرا)، دیباگر و ابزار پکیج/مستندسازی.</li>
    <li><strong>Compiler</strong> — تبدیل کد منبع <code dir = "ltr">.java</code> به بایت‌کد <code dir = "ltr">.class</code> قابل‌اجرا روی JVM.</li>
    <li><strong>Documentation</strong> — مستندات زبان و APIها برای توسعه سریع‌تر و درست‌تر.</li>
    <li><strong>JRE</strong> — شامل:
      <ul>
        <li><strong>JVM</strong> برای اجرای بایت‌کد.</li>
        <li><strong>Libraries</strong> کتابخانه‌های استاندارد برای ورودی/خروجی، شبکه، کالکشن‌ها و …</li>
      </ul>
    </li>
  </ul>
  <div class="slide-note">
    جمع‌بندی: <strong>JDK = ابزارهای توسعه + JRE</strong>، در حالی‌که <strong>JRE = فقط اجرای برنامه</strong>.
  </div>

<div class="card" dir="rtl">
  <h3>خصوصیات زبان جاوا</h3>
  <ul>
    <li>
      <strong>سادگی:</strong>
      سینتکس آشنا شبیه C اما بدون اشاره‌گر؛ مدیریت حافظه خودکار با <em>Garbage Collection</em>.
    </li>
    <li>
      <strong>شیءگرایی (Object-Oriented):</strong>
      مدل‌سازی دنیای واقعی با کلاس/شیء، تعامل بهتر اجزا، و <strong>استفاده‌مجدد از کد</strong>.
    </li>
    <div class="images">
    <img src="img/oop.jpg" alt="Java Logo">
    </div>
    <li>
      <strong>قابل‌حمل و مستقل از معماری:</strong>
      یک کد جاوا روی سامانه‌های مختلف بدون تغییر اجرا می‌شود.
    </li>
        <li>
        <strong>ترکیب کامپایل و تفسیر:</strong>
        در جاوا ابتدا کد منبع توسط <em>کامپایلر</em> به قالبی به نام <code>Bytecode</code> تبدیل می‌شود. این کد میانی مستقل از سیستم‌عامل است و توسط <strong>ماشین مجازی جاوا (JVM)</strong> در زمان اجرا تفسیر یا به کد بومی ترجمه می‌شود. این فرآیند باعث می‌شود برنامه بدون تغییر روی سیستم‌عامل‌ها و معماری‌های مختلف اجرا شود (<em>Platform-Independent</em>).
      </li>
            <div class="images">
    <img src="img/independent.png" alt="Java Logo">
    </div>
    <li>
    <strong>چندریسمانی (Multithreading):</strong>
    قابلیتی که به برنامه اجازه می‌دهد چند بخش (Thread) را به‌طور همزمان اجرا کند. این ویژگی باعث می‌شود وظایف سنگین مانند پردازش داده، رندر گرافیک، یا ارتباطات شبکه‌ای بدون ایجاد توقف در عملکرد اصلی برنامه انجام شوند. جاوا ابزارها و کلاس‌های داخلی متنوعی برای مدیریت Threadها، همگام‌سازی (Synchronization) و جلوگیری از تداخل داده‌ها فراهم می‌کند، که در نتیجه کارایی و پاسخ‌گویی برنامه بهبود می‌یابد.
  </li>
  <li>
    <strong>امن و قابل اتکا (Secure & Robust):</strong>
    مدل امنیتی قوی، بررسی خطا در زمان کامپایل/اجرا، عدم دسترسی مستقیم به حافظه.
  </li>
  </ul>
</div>

  <h3>مطالعه بیشتر: تاریخچه و کاربردهای جاوا</h3>

  <p>
    زبان <strong>جاوا</strong> در اوایل دههٔ ۹۰ میلادی در زمانی متولد شد که صنعت نرم‌افزار با یک چالش جدی روبه‌رو بود:
    هر سیستم‌عامل و سخت‌افزار زبان و محیط توسعهٔ مخصوص به خود را داشت و انتقال یک نرم‌افزار از یک پلتفرم به دیگری
    هزینه و زمان زیادی می‌برد. شرکت <strong>Sun Microsystems</strong> تصمیم گرفت زبانی طراحی کند که مستقل از سخت‌افزار
    و سیستم‌عامل باشد. نتیجهٔ این تلاش <em>ماشین مجازی جاوا (Java Virtual Machine - JVM)</em> و شعار مشهور
    <strong>«یک‌بار بنویس، همه‌جا اجرا کن»</strong> بود.  
    ایدهٔ اصلی این بود که برنامه‌ها به جای اجرا شدن مستقیم روی سیستم‌عامل، روی یک ماشین مجازی اجرا شوند که خودش با
    هر سیستم‌عامل هماهنگ می‌شود. این نوآوری باعث شد جاوا به سرعت جایگاه ویژه‌ای در توسعهٔ نرم‌افزار پیدا کند، به‌خصوص
    در پروژه‌هایی که نیاز به پایداری، امنیت و قابلیت حمل بالا داشتند.
  </p>

  <h4 style="color: #dc2626;">پیدایش (۱۹۹۱–۱۹۹۵)</h4>
  <p>
    در سال ۱۹۹۱، <strong>James Gosling</strong> به همراه تیمی از مهندسان Sun Microsystems پروژه‌ای با نام رمز
    <em>Oak</em> را آغاز کرد. هدف اولیه، ساخت زبانی سبک و امن برای دستگاه‌های تعاملی و سیستم‌های توکار (مانند ست‌تاپ‌باکس‌ها)
    بود. این زبان باید بتواند بدون وابستگی به سخت‌افزار، در محیط‌های مختلف اجرا شود.  
    با رشد اینترنت در اوایل دههٔ ۹۰، تیم توسعه متوجه شد که همین قابلیت «چندسکویی» می‌تواند برای وب نیز بسیار ارزشمند باشد.
    به همین دلیل، در سال ۱۹۹۵ این زبان با نام جدید <strong>Java</strong> معرفی شد و همراه با مرورگر <em>HotJava</em>
    و تکنولوژی اپلت (Applet) به جهان عرضه گردید.
  </p>

  <h4 style ="color: #dc2626;">رشد چشمگیر کجا رخ داد؟</h4>
  <ul>
    <li><strong>وب سازمانی (اواخر دههٔ ۹۰ و ۲۰۰۰):</strong> با معرفی <em>Servlet/JSP</em> و سپس <strong>J2EE</strong>، جاوا به هستهٔ اصلی سیستم‌های سازمانی در بانکداری، مخابرات و تجارت الکترونیک تبدیل شد.</li>
    <li><strong>اندروید (از ۲۰۰۸):</strong> عرضهٔ Android باعث شد میلیون‌ها توسعه‌دهنده وارد اکوسیستم جاوا شوند و برنامه‌های موبایل خود را با این زبان بنویسند.</li>
    <li><strong>زیرساخت‌های مقیاس‌پذیر:</strong> فریم‌ورک‌هایی مانند <strong>Spring Boot</strong> و فناوری‌های مبتنی بر JVM (مثل Kafka و Hadoop) راه را برای توسعه سیستم‌های ابری و مقیاس‌پذیر هموار کردند.</li>
  </ul>
